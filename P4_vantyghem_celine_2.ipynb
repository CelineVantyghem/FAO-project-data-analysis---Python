utlisation des df dispo_alimentaire, population, comparaison des listes "Zone", corrections, création des deux tables pivot, merge, calcul de la disponibilité alimentaire totale(kcal/pers/jour/pays * population, division par l'apport nécessaire en calories par personne par jour

```python
nbr_calorie_needed_perperson_perday = 2400
```

```python
dispo_alimentaire.head()
```

```python
#méthode groupby :
#dispo_alimentaire.groupby("Zone")["Disponibilité alimentaire (Kcal/personne/jour)"].sum()
```

```python
#méthode pivot sur le df disponibilité alimentaire : 
calories_2013_perperson_perday_perzone_pivot = dispo_alimentaire.pivot_table(values = "Disponibilité alimentaire (Kcal/personne/jour)", index = "Zone", aggfunc=sum)
calories_2013_perperson_perday_perzone_pivot
```

```python
calories_2013_perperson_perday_perzone_pivot.info()
```

```python
#réduction df population sur les valeurs 2017 :
is_2017_population = population["Année"] == 2017
population_2017 =  population[is_2017_population]
population_2017.head()
```

```python
#comparaison des valeurs "Zones" entre le df disponibilité alimentaire et le df population :
liste_zones_dispo_alimentaire = calories_2013_perperson_perday_perzone_pivot.index.tolist()
```

```python
liste_zone_population_2017 = population_2017["Zone"].tolist()
```

```python
#il y a un plus grand nombre de valeurs "Zone" dans le df population que dans le df disponibilité alimentaire. Réduction du nombre de valeurs "Zone" du df population pour correspondre aux valeurs du df disponibilité alimentaire :
is_population_2017_reduction_zones = population_2017["Zone"].isin(liste_zones_dispo_alimentaire)
population_2017_reduction_zones= population_2017[is_population_2017_reduction_zones]
zones_population_2017_reduit = population_2017_reduction_zones["Zone"]
liste_zones_population_2017_reduit = zones_population_2017_reduit.tolist()
len(liste_zones_population_2017_reduit)
```

```python
#174 vs 172 : 2 pays de la liste "liste_zones_dispo_alimentaire" n'ont pas été retrouvés, lesquels ? : 
set(liste_zones_dispo_alimentaire) - set(liste_zones_population_2017_reduit)
```

```python
#correction des différence orthographiques : 
population_2017 = population_2017.replace(to_replace = "Royaume-Uni de Grande-Bretagne et d'Irlande du Nord", value="Royaume-Uni" )
population_2017 = population_2017.replace(to_replace = "Tchéquie", value="Tchéquie (la)" )
```

```python
#vérification des corrections : 
liste_zone_population2 = population_2017["Zone"].tolist()
```

```python
#réduction du nombre de valeurs "Zone" du df population modifié pour correspondre aux valeurs du df disponibilité alimentaire :
is_population_2017_reduction_zones = population_2017["Zone"].isin(liste_zones_dispo_alimentaire)
population_2017_reduction_zones= population_2017[is_population_2017_reduction_zones]
```

```python
zones_population_2017_reduit = population_2017_reduction_zones["Zone"]
```

```python
#il faut également modifier le nombre de categories (ici 236) afin de pouvoir effecter un merge (mêmes catégories requises entre les 2 df) :
zones_population_2017_reduit= zones_population_2017_reduit.cat.remove_unused_categories()
```

```python
#création d'un dataframe avec le nouveau series zones_population_2017_reduit :
population_2017_reduction_zones_cat =  pd.concat([zones_population_2017_reduit, population_2017_reduction_zones['Valeur'] ], axis=1)
```

```python
#méthode pivot sur le df population_2017_reduction_zones_cat :
population_2017_reduction_zones_pivot = population_2017_reduction_zones_cat.pivot_table(values = "Valeur", index = "Zone")
population_2017_reduction_zones_pivot
```

```python
#fusion des deux tables pivot population 2017 et calories 2013 :
total_calories_perday_percountry_merge = pd.merge(calories_2013_perperson_perday_perzone_pivot, population_2017_reduction_zones_pivot, on = "Zone")
total_calories_perday_percountry_merge
```

```python
#création de la colonne "Calories par jour par pays" :
total_calories_perday_percountry_merge["Calories par jour par pays"] = total_calories_perday_percountry_merge["Valeur"]*total_calories_perday_percountry_merge["Disponibilité alimentaire (Kcal/personne/jour)"]
total_calories_perday_percountry_merge
```

```python
total_calories = total_calories_perday_percountry_merge["Calories par jour par pays"].sum()
total_calories
```

```python
nbr_theorique_personnes_nourries = round(total_calories/nbr_calorie_needed_perperson_perday)
print(f'Nombre théorique de personnes pouvant être nourries : {nbr_theorique_personnes_nourries}')
```

```python
total_population_2017 = population_2017["Valeur"].sum()
proportion_theorique_personnes_nourries = nbr_theorique_personnes_nourries/total_population_2017
print(f'Proportion théorique de la population pouvant être nourrie: {round(proportion_theorique_personnes_nourries,2)*100}%')
```
